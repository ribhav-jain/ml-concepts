```python
import pandas as pd

messages = pd.read_csv(
    "SpamClassifier/SMSSpamCollection", sep="\t", names=["label", "message"]
)
```

```python
messages.shape
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
import re
import nltk

nltk.download("stopwords")
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub("[^a-zA-Z]", " ", messages["message"][i])
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review]
    review = " ".join(review)
    corpus.append(review)
```

```python
[
    [i, j, k]
    for i, j, k in zip(list(map(len, corpus)), corpus, messages["message"])
    if i < 1
]
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
from gensim.models import Word2Vec
```

```python
## Lets train Word2vec from scratch
model = gensim.models.Word2Vec(words)
```

```python
# Step 2: Train the model
model.train(words, total_examples=len(words), epochs=model.epochs)
```

```python
# tqdm is used to display a smart progress bar for loops
# Great for tracking progress in long-running tasks
# Example: for i in tqdm(range(100)): ...
from tqdm import tqdm
```

```python
# Updated average word2vec function
def avg_word2vec(tokens, model, vector_size=100):
    valid_vectors = [model.wv[word] for word in tokens if word in model.wv]

    if valid_vectors:
        return np.mean(valid_vectors, axis=0)
    else:
        return np.zeros(vector_size)  # Fallback for unknown/empty tokens


# Apply it to all sentences
X = []
for i in tqdm(range(len(words))):  # `words` is a list of tokenized sentences
    X.append(avg_word2vec(words[i], model, vector_size=100))  # pass model + vector_size

# Convert to a NumPy array safely
X_new = np.array(X)
```
